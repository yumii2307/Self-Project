```python
import pandas as pd
```

```python
CCTV_Seoul = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황.csv', encoding='euc-kr', thousands = ',')
CCTV_Seoul.head()
```

```python
CCTV_Seoul = CCTV_Seoul[['구분', '총계', '2013년 이전\n설치된 CCTV', '2013년', '2014년', '2015년', '2016년']]
CCTV_Seoul.head()
```

```python
CCTV_Seoul.columns = ['구분', '총계', '2013년 이전', '2013년', '2014년', '2015년', '2016년']
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_csv('data/서울시 주민등록인구.csv', skiprows=1)
pop_Seoul.head()
```

```python
pop_Seoul = pop_Seoul[['동별(2)', '계', '한국인', '등록외국인', '65세이상고령자']]
pop_Seoul.head()
```

```python
pop_Seoul.columns = ['구분', '총계', '내국인', '외국인', '고령자']
pop_Seoul.head()
```

```python
CCTV_Seoul = CCTV_Seoul[CCTV_Seoul.구분 != '계']
CCTV_Seoul = CCTV_Seoul.astype({'2013년 이전':int, '2013년':int, '2014년':int, '2015년':int, '2016년':int})
CCTV_Seoul.head()
```

```python
CCTV_Seoul['총계'] = CCTV_Seoul.iloc[:, 2:].sum(axis=1)
CCTV_Seoul.head()
```

```python
CCTV_Seoul.sort_values(by='총계', ascending=True).head(5)
```

```python
CCTV_Seoul.sort_values(by='총계', ascending=False).head(5)
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년'] + CCTV_Seoul['2013년'] / CCTV_Seoul['2013년 이전'] * 100)
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
pop_Seoul.drop([0], inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul['구분'].unique()
```

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['총계'] * 100
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['총계'] * 100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by='총계', ascending=False).head()
```

```python
pop_Seoul.sort_values(by='외국인', ascending=False).head()
```

```python
pop_Seoul.sort_values(by='외국인비율', ascending=False).head()
```

```python
pop_Seoul.sort_values(by='고령자', ascending=False).head()
```

```python
pop_Seoul.sort_values(by='고령자비율', ascending=False).head()
```

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on='구분')
data_result.head()
```

```python
del data_result['2013년 이전']
del data_result['2013년']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
```

```python
data_result.columns = ['구분', '총계', '최근증가율', '인구수', '내국인', '외국인', '고령자', '외국인비율', '고령자비율']
data_result.head()
```

```python
data_result.set_index('구분', inplace=True)
data_result.head()
```
